# Matplotlib Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
#Use Numpy to generate random data
import numpy as np
x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
#It only take 1 command to draw
pylab.plot(x,y, 'r') #'r' stands for draw in red
```

```python
#Drawing a subgraph
pylab.subplot(1,2,1)#The content of the brackets represents rows, colums and indexes
pylab.plot(x,y, 'r--')#Third parameter is color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')
```

```python
#(1,2,1) and (1,2,2) are signficant if we want tiled subplot, without which both the grap[hs will be in the same plot
#For example
pylab.subplot(1,2,1)#The content of the brackets represents rows, colums and indexes
pylab.plot(x,y, 'r--')#Third parameter is color and line style
pylab.subplot(1,2,1)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) #control the left, right, width, height of the canvsa (from 0 to 1)

axis.plot(x,y, 'r')
```

```python
#Again we can draw subgraphs
fig, axes = plt.subplots(nrows = 1, ncols=2) #submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
#We can also draw a picture or graph inside another graph
fig = plt.figure() #our convas

#control the left, bottom, width, height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1,0.,0.8,0.8]) #big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #small convas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize=(16,9), dpi=250) #New graphic object

fig.add_subplot()

plt.plot(x, y, 'r')
```

```python
ax.legend(["label1", "label2"])

fig, axes = plt.subplots(dpi=200)
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")

axes.set_title("My Chart")

axes.plot(x, x**2)
axes.plot(x, x**3)
          
axes.legend(["y = x**2", "y = x**3"], loc=2)
```

```python
#In matplotlib, you can set other parameter such as line color, transparenncy etc.
fig, axes = plt.subplots(dpi=150)
axes.plot(x, x**2, color='red', alpha=.5)
axes.plot(x, x+2, color="#1155dd", alpha=.5)
axes.plot(x, x+3, color="#15cc55", alpha=.5)
```

```python
fig, ax = plt.subplots(dpi=100)

#line width
ax.plot(x, x+1, color='blue', linewidth=0.25)
ax.plot(x, x+2, color='blue', linewidth=1)
ax.plot(x, x+3, color='blue', linewidth=1.5)
ax.plot(x, x+4, color='blue', linewidth=2)
```

```python
fig, ax = plt.subplots(dpi=100)

#line width
ax.plot(x, x+1, color='blue', lw=2, linestyle='-')
ax.plot(x, x+2, color='blue', lw=2, linestyle='-.')
ax.plot(x, x+3, color='blue', lw=2, linestyle=':')

line, = ax.plot(x, x+4, color='black', lw=1.50)
line.set_dashes([5,10,15,10])

line, = ax.plot(x, x+5, color='black', lw=1.50)
line.set_dashes([5,10,15,3])

line, = ax.plot(x, x+6, color='black', lw=1.50)
line.set_dashes([5,10,4,10])
```
